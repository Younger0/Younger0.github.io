```python
snp = yf.Ticker("^GSPC").history(period='10y')
ftse = yf.Ticker("^FTSE").history(period='10y')
```

```python
snp=snp[['Close']]
snp['Name']='S&P500'
```

```python
snp
```

```python
ftse=ftse[['Close']]
ftse['Name']='FTSE100'
```

```python
ftse
```

```python
dfs=[]
for x in ['^GSPC','^FTSE']:
  df = yf.Ticker(x).history(period='10y')
  df=df[['Close']]
  df['Name']=x
  dfs.append(df) 
```

```python
df
```

```python
import pandas as pd
pd.concat([ dfs[0],dfs[1] ])
```

```python
dfz=pd.DataFrame() #initialise empty DataFrame
for x in dfs:
  stock_name=x['Name'].values[0]
  stock_name=stock_name.replace('.','')
  x=x[['Close']]
  x.columns=[stock_name]
  dfz=x.join(dfz)
```

```python
dfz
```

```python
dfz.dropna(subset=["^FTSE", "^GSPC"], inplace=True)
```
